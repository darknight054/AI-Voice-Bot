```python
from transformers import pipeline
from transformers.utils import is_flash_attn_2_available
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, TextIteratorStreamer, TextStreamer, set_seed, AutoFeatureExtractor
from parler_tts import ParlerTTSForConditionalGeneration
from typing import Optional
import io
import math
from queue import Queue
from threading import Thread
import numpy as np
import gradio as gr
import scipy.io.wavfile as sw
import os
from transformers.pipelines.audio_utils import ffmpeg_read
from pydub import AudioSegment
from parler_tts.streamer import ParlerTTSStreamer
import string
import uuid
import time

device = "cuda:0" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu" # set the device
attn_implementation = "flash_attention_2" if is_flash_attn_2_available() else "sdpa" # T4s do not support flash_attention_2, Using SDPA
```

TTS Model: https://huggingface.co/parler-tts/parler-tts-mini-v1

```python
# Import the text to speech model and respective tokenizer
tts_model_name = "parler-tts/parler_tts_mini_v0.1"
description = "very clear audio"
tts_model = ParlerTTSForConditionalGeneration.from_pretrained(
    tts_model_name, torch_dtype=torch.float16, low_cpu_mem_usage=True).to(device)
tts_tokenizer = AutoTokenizer.from_pretrained(tts_model_name)
tts_sampling_rate = tts_model.audio_encoder.config.sampling_rate
tts_frame_rate = tts_model.audio_encoder.config.frame_rate
# Compile the model using torch.compile for some performance benefits. For the demo, avoiding it.
# compile_mode = "reduce-overhead"
# tts_model.generation_config.cache_implementation = "static"
# tts_model.forward = torch.compile(tts_model.forward, mode=compile_mode)

# # warmup
# inputs = tts_tokenizer("This is for compilation", return_tensors="pt", padding="max_length", max_length=50).to(device)

# model_kwargs = {**inputs, "prompt_input_ids": inputs.input_ids, "prompt_attention_mask": inputs.attention_mask, }

# n_steps = 1 if compile_mode == "default" else 2
# for _ in range(n_steps):
#     _ = tts_model.generate(**model_kwargs)
```

Speech to Text Model: https://huggingface.co/openai/whisper-tiny.en

```python
# Import the speech to text model and respective tokenizer (using huggingface pipelines for ease of use and it supports chunking as well)
stt_model_name = "openai/whisper-tiny.en"
stt_pipe = pipeline(
    "automatic-speech-recognition",
    model=stt_model_name,
    torch_dtype=torch.bfloat16, # load in bfloat16 for more accuracy
    device=device,
    chunk_length_s=30,
    model_kwargs={"attn_implementation": attn_implementation},
)
```

LLM model: https://huggingface.co/TheBloke/TinyLlama-1.1B-Chat-v1.0-AWQ

```python
# Importing the LLM model and Tokenizer in the AWQ quanitsized version to save GPU memory (Complete model could also be loaded)
llm_model_name = "TheBloke/TinyLlama-1.1B-Chat-v1.0-AWQ"
llm_tokenizer = AutoTokenizer.from_pretrained(llm_model_name)
llm_model = AutoModelForCausalLM.from_pretrained(
    llm_model_name,
    low_cpu_mem_usage=True,
    device_map=device,
    attn_implementation=attn_implementation
)
# Defining the Streamer
llm_streamer = TextIteratorStreamer(llm_tokenizer, skip_prompt=True, skip_special_tokens=True)
# Defining the System prompt
llm_system_prompt = f"""<|system|>You are a helpful transcribing assistant</s>"""
```

```python
# Function to convert numpy array to a WAV file
def save_numpy_to_wav(audio_np, sample_rate, filename):
    sw.write(filename, sample_rate, audio_np)
    print(f"Audio written to disk with filename: {filename}")

def remove_punctuation(input_string):
    # Remove punctuation from the input string
    return input_string.translate(str.maketrans('', '', string.punctuation))

def add_message(history, audio_input):
    # Append the user's audio input as a message (displayed as an audio file) in the chat history
    history.append((gr.Audio(audio_input,elem_classes="custom-audio", editable=False), None))
    return history

def numpy_to_mp3(audio_array, sampling_rate):
    # Normalize audio_array if it's floating-point
    if np.issubdtype(audio_array.dtype, np.floating):
        max_val = np.max(np.abs(audio_array))
        audio_array = (audio_array / max_val) * 32767  # Normalize to 16-bit range
        audio_array = audio_array.astype(np.int16)

    # Create an audio segment from the numpy array
    audio_segment = AudioSegment(
        audio_array.tobytes(),
        frame_rate=sampling_rate,
        sample_width=audio_array.dtype.itemsize,
        channels=1
    )

    # Export the audio segment to MP3 bytes - use a high bitrate to maximise quality
    mp3_io = io.BytesIO()
    audio_segment.export(mp3_io, format="mp3", bitrate="320k")

    # Get the MP3 bytes
    mp3_bytes = mp3_io.getvalue()
    mp3_io.close()

    return mp3_bytes

# Implements the logic of Voice Bot Response
def bot_response(audio_input, llm_generation_kwargs, chat_history_state, play_steps_in_s=5.0, description="very clear audio"):

    # Open the audio input file and read it in binary format
    with open(audio_input, "rb") as f:
        audio_input = f.read()

    # Convert the audio input into a format acceptable for the speech-to-text (STT) pipeline
    audio_input = ffmpeg_read(audio_input, stt_pipe.feature_extractor.sampling_rate)
    stt_input = {"sampling_rate": stt_pipe.feature_extractor.sampling_rate, "raw": audio_input}

    # Perform speech-to-text conversion using the STT pipeline
    stt_text = stt_pipe(stt_input).get("text")

    # If STT successfully transcribes the audio
    if stt_text:
        print("STT Text: ", stt_text)

        # Create a user prompt template using the transcribed text
        user_prompt_template = f"""<|user|>{stt_text}</s><|assistant|>"""

        # Update the chat history state with the new prompt
        chat_history_state += user_prompt_template

        # Tokenize the chat history state and prepare generation arguments for the LLM
        llm_tokens = llm_tokenizer(chat_history_state, return_tensors='pt').to(device)
        generation_kwargs = dict(llm_tokens, streamer=llm_streamer) | llm_generation_kwargs

        # Debug: print the generation arguments to verify them
        print(generation_kwargs)

        # Set a seed for reproducibility
        set_seed(42)

        # Start a separate thread to handle the LLM generation process
        thread = Thread(target=llm_model.generate, kwargs=generation_kwargs)
        thread.start()

        # Initialize an empty string to store the generated text
        generated_text = ""

        # Continuously collect the text output from the LLM streamer as it's generated
        for new_text in llm_streamer:
            generated_text += new_text

        # Update the chat history with the newly generated text
        chat_history_state += f"""{generated_text}</s>"""
        print("LLM Text Output: ", generated_text)

        # Optionally remove punctuation for better downstream results (getting better results)
        generated_text = remove_punctuation(generated_text)
        print("Chat History: ", chat_history_state)

        # Calculate the number of steps for Text-to-Speech (TTS) playback
        play_steps = int(tts_frame_rate * play_steps_in_s)

        # Initialize the TTS streamer with the given play steps
        tts_streamer = ParlerTTSStreamer(tts_model, device=device, play_steps=play_steps)

        # Tokenize the TTS description input and prompt
        description_inputs = tts_tokenizer(description, return_tensors="pt").to(device)
        tts_prompt = tts_tokenizer(generated_text, return_tensors="pt").to(device)

        # Prepare generation arguments for the TTS model
        tts_generation_kwargs = dict(
            input_ids=description_inputs.input_ids,
            prompt_input_ids=tts_prompt.input_ids,
            streamer=tts_streamer,
            do_sample=True,
            temperature=1.0,  # Add randomness to the TTS output
            min_new_tokens=10  # Ensure a minimum number of tokens in the generated audio
        )

        # Set a seed for TTS generation reproducibility
        set_seed(42)

        # Start a new thread to generate audio from the TTS model
        thread = Thread(target=tts_model.generate, kwargs=tts_generation_kwargs)
        thread.start()

        # Initialize an empty array to store the complete audio data
        complete_audio = np.array([])
        filename = str(uuid.uuid4()) + ".wav"
        # Collect generated audio in chunks from the TTS streamer
        for new_audio in tts_streamer:
            # Log the length of each audio sample in seconds
            print(f"Sample of length: {round(new_audio.shape[0] / tts_sampling_rate, 2)} seconds")

            # Yield each audio chunk as MP3 to the client
            yield numpy_to_mp3(new_audio, sampling_rate=tts_sampling_rate)

            # Concatenate the new audio chunk to the complete audio array
            complete_audio = np.concatenate((complete_audio, new_audio))

        # Save the complete audio to a .wav file in a separate thread
        thread = Thread(target=save_numpy_to_wav, args=(complete_audio, tts_sampling_rate, filename))
        thread.start()

# Gradio UI with numpy array input and button enable/disable logic
def gradio_ui():
    # Create a Gradio Blocks interface for the app
    with gr.Blocks() as app:
        gr.Markdown("# General Purpose Voice bot")  # Display a title for the UI

        # Set up the layout with two columns: Chatbot UI and Audio Output
        with gr.Row(equal_height=False):
          with gr.Column():
            chatbot_ui = gr.Chatbot(scale=1, elem_id="chatbot", container=False)  # Chatbot panel
          with gr.Column():
            audio_out = gr.Audio(label="Bot Output", format="mp3", elem_id="audio_out",
                                 interactive=False, streaming=True, autoplay=True,
                                 scale=1, waveform_options={"show_recording_waveform": True, "show_controls": True})  # Audio output player

        # Row containing audio input and submit button
        gr.Markdown("Please wait for the audio to be loaded, Submit button will be enabled automatically post that. This lag is due to Gradio Audio Component.<br> Use 3-5 seconds of input audio for ease of usage")
        with gr.Row():
            audio_input = gr.Audio(
                sources=["microphone"],  # Only accept input from the microphone
                type="filepath",  # Return the path to the audio file
                label="Record Your Voice",  # Label for the audio input
                format="wav",  # Input format is WAV
                interactive=True  # Allow interaction with the input
            )

            # Initially disable the submit button until valid input is provided
            submit_button = gr.Button("Submit", interactive=False)

        # Tab for configuring LLM parameters
        with gr.Tab("LLM Parameters"):
          with gr.Row():
            temperature = gr.Slider(
                label="Temperature", value=0.7, minimum=0.0, maximum=1.0, step=0.05,
                interactive=True, info="Higher values produce more diverse outputs"
            )
            max_new_tokens = gr.Slider(
                label="Max new tokens", value=32, minimum=0, maximum=2048, step=64,
                interactive=True, info="Maximum number of new tokens generated"
            )
            top_p = gr.Slider(
                label="Top-p (nucleus sampling)", value=0.95, minimum=0.0, maximum=1.0,
                step=0.05, interactive=True, info="Samples from more low-probability tokens with higher values"
            )
            repetition_penalty = gr.Slider(
                label="Repetition penalty", value=1.1, minimum=1.0, maximum=2.0, step=0.05,
                interactive=True, info="Penalize repeated tokens in the output"
            )

        # Tab for configuring Text-to-Speech (TTS) parameters
        with gr.Tab("TTS Parameters"):
          with gr.Row():
            description = gr.Textbox(value="very clear audio", label="Optional Description", interactive=True)
            play_seconds = gr.Slider(3.0, 7.0, value=5.0, step=2, label="Streaming interval in seconds",
                                     info="Lower = shorter chunks, lower latency, but more codec steps")

        # Function to collect generation parameters for the LLM
        def collect_generation_kwargs(temperature, max_new_tokens, top_p, repetition_penalty):
            generation_kwargs = {
                "do_sample": True,
                "top_k": 40,  # Limit token sampling to the top 40 most probable tokens
                "temperature": temperature,
                "max_new_tokens": max_new_tokens,
                "top_p": top_p,
                "repetition_penalty": repetition_penalty,
            }
            return generation_kwargs

        # Function to enable the submit button when audio is available
        def enable_submit(audio):
            if audio is not None:
                return gr.update(interactive=True)  # Enable the button
            return gr.update(interactive=False)  # Keep the button disabled

        # Gradio State to maintain LLM generation arguments and chat history
        generation_kwargs_state = gr.State({})
        chat_history_state = gr.State(llm_system_prompt)  # Initialize chat history with system prompt

        # Update generation kwargs based on the user input
        def update_generation_kwargs(temp, mnt, tp, rp, state):
            generation_kwargs = collect_generation_kwargs(temp, mnt, tp, rp)
            state.update(generation_kwargs)  # Update the state with new generation parameters
            return state

        # Handle user audio input and generate bot response
        submit_button.click(
            add_message,  # Add the user's message to the chatbot UI
            inputs=[chatbot_ui, audio_input],  # Inputs: chat UI and recorded audio
            outputs=[chatbot_ui],  # Output: updated chat UI
        ).then(
            update_generation_kwargs,  # Update generation parameters after the message is added
            inputs=[temperature, max_new_tokens, top_p, repetition_penalty, generation_kwargs_state],
            outputs=[generation_kwargs_state]  # Update the state for generation kwargs
        ).then(
            bot_response,  # Generate bot response using the LLM and TTS
            inputs=[audio_input, generation_kwargs_state, chat_history_state, play_seconds, description],
            outputs=[audio_out], queue=True  # Output bot response audio
        )

        # Enable the submit button only when a valid audio file is provided
        audio_input.change(enable_submit, inputs=[audio_input], outputs=[submit_button])

    return app  # Return the configured Gradio Blocks app


# Launch the Gradio app
app = gradio_ui()
app.queue()
app.launch(share=True, debug=True)
```
